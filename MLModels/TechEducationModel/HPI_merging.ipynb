```python
import pandas as pd
import numpy as np
import cpi
```

```python
acs2017=pd.read_csv('ACS_13-17.csv',encoding="latin-1",skiprows=1)
acs2017['Year']="2013-2017"
acs2012=pd.read_csv('ACS_08-12.csv',encoding="latin-1",skiprows=1)
acs2012["Year"]="2008-2012"

acsdata=pd.merge(acs2017,acs2012,how="outer")
acsdata=acsdata.rename(columns={
    "Id2":"FIPS",
    "Estimate; Total:":"Total_pop"
})

acsdata = acsdata[acsdata.columns.drop(list(acsdata.filter(regex='Margin of Error')))]
acsdata.count()
acs=acsdata.set_index('FIPS')
acs=acs.iloc[:,2:40]

acs.columns = [col.replace('Estimate; Total: - ', 'Total: ') for col in acs.columns]
acs.columns = [col.replace('Less than high school graduate', 'less_HS') for col in acs.columns]
acs.columns = [col.replace('High school graduate (includes equivalency)', 'HS') for col in acs.columns]
acs.columns = [col.replace("Some college or associate's degree", 'some_college') for col in acs.columns]
acs.columns = [col.replace("Bachelor's degree", 'Bach') for col in acs.columns]
acs.columns = [col.replace("Graduate or professional degree", 'Grad') for col in acs.columns]
acs.columns = [col.replace("Total: ", 'Total_') for col in acs.columns]

acs["Total_Settled"]=acs["Total_Same house 1 year ago:"]+acs["Total_Moved within same county:"]
acs["Settled_less_HS"]=acs["Total_Same house 1 year ago: - less_HS"]+acs["Total_Moved within same county: - less_HS"]
acs["Settled_HS"]=acs["Total_Same house 1 year ago: - HS"]+acs["Total_Moved within same county: - HS"]
acs["Settled_some_college"]=acs["Total_Same house 1 year ago: - some_college"]+acs["Total_Moved within same county: - some_college"]
acs["Settled_Bach"]=acs["Total_Same house 1 year ago: - Bach"]+acs["Total_Moved within same county: - Bach"]
acs["Settled_Grad"]=acs["Total_Same house 1 year ago: - Grad"]+acs["Total_Moved within same county: - Grad"]
acs["Total_Moved"]=acs["Total_Moved from different county within same state:"]+acs["Total_Moved from different state:"]+acs["Total_Moved from abroad:"]
acs["Moved_less_HS"]=acs["Total_Moved from different county within same state: - less_HS"]+acs["Total_Moved from different state: - less_HS"]+acs["Total_Moved from abroad: - less_HS"]
acs["Moved_HS"]=acs["Total_Moved from different county within same state: - HS"]+acs["Total_Moved from different state: - HS"]+acs["Total_Moved from abroad: - HS"]
acs["Moved_some_college"]=acs["Total_Moved from different county within same state: - some_college"]+acs["Total_Moved from different state: - some_college"]+acs["Total_Moved from abroad: - some_college"]
acs["Moved_Bach"]=acs["Total_Moved from different county within same state: - Bach"]+acs["Total_Moved from different state: - Bach"]+acs["Total_Moved from abroad: - Bach"]
acs["Moved_Grad"]=acs["Total_Moved from different county within same state: - Grad"]+acs["Total_Moved from different state: - Grad"]+acs["Total_Moved from abroad: - Grad"]

acs=acs.drop(columns={
    "Total_Same house 1 year ago:",
    "Total_Moved within same county:",
    "Total_Same house 1 year ago: - less_HS",
    "Total_Moved within same county: - less_HS",
    "Total_Same house 1 year ago: - HS",
    "Total_Moved within same county: - HS",
    "Total_Same house 1 year ago: - some_college",
    "Total_Moved within same county: - some_college",
    "Total_Same house 1 year ago: - Bach",
    "Total_Moved within same county: - Bach",
    "Total_Same house 1 year ago: - Grad",
    "Total_Moved within same county: - Grad",
    "Total_Moved from different county within same state:",
    "Total_Moved from different state:",
    "Total_Moved from abroad:",
    "Total_Moved from different county within same state: - less_HS",
    "Total_Moved from different state: - less_HS",
    "Total_Moved from abroad: - less_HS",
    "Total_Moved from different county within same state: - HS",
    "Total_Moved from different state: - HS",
    "Total_Moved from abroad: - HS",
    "Total_Moved from different county within same state: - some_college",
    "Total_Moved from different state: - some_college",
    "Total_Moved from abroad: - some_college",
    "Total_Moved from different county within same state: - Bach",
    "Total_Moved from different state: - Bach",
    "Total_Moved from abroad: - Bach"
})
acs=acs.drop(columns={
    "Total_Moved from different county within same state: - Grad",
    "Total_Moved from different state: - Grad",
    "Total_Moved from abroad: - Grad"
})
acs.head()
```

```python
new = pd.read_csv('hpi_county.csv')
new = new.rename(columns={
    'FIPS code':'FIPS',
    "HPI with 1990 base":"HPI_1990",
    "HPI with 2000 base":"HPI_2000",
    "Annual Change (%)":"Ann_Change_%"
})
new=new.loc[new["HPI"]!="."]
new=new.loc[new["HPI_1990"]!="."]
new=new.loc[new["HPI_2000"]!="."]
new=new.loc[new["Ann_Change_%"]!="."]
new['HPI_1990']=new['HPI_1990'].replace('.', None).astype(float)
new['HPI_2000']=new['HPI_2000'].replace('.', None).astype(float)
new['Ann_Change_%']=new['Ann_Change_%'].replace('.', None).astype(float)
new['HPI']=new['HPI'].replace('.', None).astype(float)
new['FIPS']=new['FIPS'].astype(object)
new.head()
```

```python
new['HPI_adj'] = new.apply(lambda x: cpi.inflate(x.HPI, x.Year), axis=1)
new['HPI_2000_adj'] = new.apply(lambda x: cpi.inflate(x.HPI_2000, x.Year), axis=1)
new['HPI_1990_adj'] = new.apply(lambda x: cpi.inflate(x.HPI_1990, x.Year), axis=1)
```

```python
melted_HPI = pd.melt(new, 
        id_vars=['State','County','FIPS','Year'], 
        value_vars=['Ann_Change_%', 'HPI', 'HPI_1990',
       'HPI_2000', 'HPI_adj', 'HPI_2000_adj', 'HPI_1990_adj'])
# melted_HPI = melted_HPI.loc[(melted_HPI["Year"]>2007)&(melted_HPI["Year"]<2018)]
```

```python
melted_HPI = pd.merge(melted_HPI,acs,how="left",on="FIPS")
melted_HPI.head()
```

```python
binned_df = new.loc[(new["Year"]>2007)&(new["Year"]<2018)]
binned_df=binned_df.groupby([binned_df.FIPS,binned_df.County,binned_df.State,pd.cut(binned_df.Year,[2007,2012,2017]).astype(str)])[['HPI','HPI_1990','HPI_2000','Ann_Change_%','HPI_adj','HPI_1990_adj','HPI_2000_adj']].mean().reset_index()
binned_df['Year_Bin'] = np.where(binned_df['Year']=="(2007, 2012]", '2008-2012', '2013-2017')
```

```python
merged = pd.merge(binned_df,acs,how='left',left_on=['Year_Bin',"FIPS"],right_on=["Year",'FIPS'])
```

```python
merged["Years_Education_Settled"]=\
    (merged["Settled_less_HS"]/merged["Total_Settled"])*11+\
    (merged["Settled_HS"]/merged["Total_Settled"])*12+\
    (merged["Settled_some_college"]/merged["Total_Settled"])*14+\
    (merged["Settled_Bach"]/merged["Total_Settled"])*16+\
    (merged["Settled_Grad"]/merged["Total_Settled"])*22

merged["Years_Education_Moved"]=\
    (merged["Moved_less_HS"]/merged["Total_Moved"])*11+\
    (merged["Moved_HS"]/merged["Total_Moved"])*12+\
    (merged["Moved_some_college"]/merged["Total_Moved"])*14+\
    (merged["Moved_Bach"]/merged["Total_Moved"])*16+\
    (merged["Moved_Grad"]/merged["Total_Moved"])*22

merged["Moved_Ratio"]=merged["Total_Moved"]/merged["Total_pop"]
merged["Moved_Settled_Ratio"]=merged["Total_Moved"]/merged["Total_Settled"]
merged["Grad_Bach_Ratio"]=merged["Total_Grad"]/merged["Total_Bach"]
merged["College_Ratio"]=(merged["Total_Bach"]+merged["Total_Grad"])/merged["Total_pop"]
merged["Grad_Ratio_Moved"]=merged["Moved_Grad"]/merged["Total_Moved"]
merged["Years_Edu_Moved_Minus_Settled"]=merged["Years_Education_Moved"]-merged["Years_Education_Settled"]
merged["Total_Grad_Ratio"]=merged["Total_Grad"]/merged["Total_pop"]
merged["Total_Bach_Ratio"]=merged["Total_Bach"]/merged["Total_pop"]
merged.head()
```

```python
tech=pd.read_csv('tech_industry.csv',encoding="latin-1",skiprows=1)
tech.columns
```

```python
totals=tech.loc[tech["2012 NAICS code"]==0]
totals=totals.rename(columns={
    "Paid employees for pay period including March 12 (number)":"Total_Employees",
    "Id2":"FIPS",
    "Number of establishments":"Total_Establishments"
})
totals=totals.drop(columns={
    "Geographic area name","Geographic identifier code","Id","2012 NAICS code","Meaning of 2012 NAICS code","Year","Id.1","First-quarter payroll ($1,000)","Annual payroll ($1,000)"
})
totals.head()
```

```python
counties=tech.loc[tech["2012 NAICS code"]!=0]
counties=counties.groupby(by="Id2").sum()

counties=counties.drop(columns={
    "Id","2012 NAICS code","Id.1","Year","First-quarter payroll ($1,000)","Annual payroll ($1,000)"
})
counties=counties.reset_index()
counties=counties.rename(columns={
    "Paid employees for pay period including March 12 (number)":"Tech_Employees",
        "Id2":"FIPS",
    "Number of establishments":"Tech_Establishments"
})
counties.head()
```

```python
tech_industry = pd.merge(totals,counties,how="left",on="FIPS")
tech_industry.head()
tech_industry["Tech_Establishments_Share"]=tech_industry["Tech_Establishments"]/tech_industry["Total_Establishments"]
tech_industry["Tech_Employees_Share"]=tech_industry["Tech_Employees"]/tech_industry["Total_Employees"]
```

```python
melted_HPI=pd.merge(melted_HPI,tech_industry,how="left",on="FIPS")
melted_HPI.head()
```

```python
melted_HPI["Years_Education_Settled"]=\
    (melted_HPI["Settled_less_HS"]/melted_HPI["Total_Settled"])*11+\
    (melted_HPI["Settled_HS"]/melted_HPI["Total_Settled"])*12+\
    (melted_HPI["Settled_some_college"]/melted_HPI["Total_Settled"])*14+\
    (melted_HPI["Settled_Bach"]/melted_HPI["Total_Settled"])*16+\
    (melted_HPI["Settled_Grad"]/melted_HPI["Total_Settled"])*22

melted_HPI["Years_Education_Moved"]=\
    (melted_HPI["Moved_less_HS"]/melted_HPI["Total_Moved"])*11+\
    (melted_HPI["Moved_HS"]/melted_HPI["Total_Moved"])*12+\
    (melted_HPI["Moved_some_college"]/melted_HPI["Total_Moved"])*14+\
    (melted_HPI["Moved_Bach"]/melted_HPI["Total_Moved"])*16+\
    (melted_HPI["Moved_Grad"]/melted_HPI["Total_Moved"])*22

melted_HPI["Moved_Ratio"]=melted_HPI["Total_Moved"]/melted_HPI["Total_pop"]
melted_HPI["Moved_Settled_Ratio"]=melted_HPI["Total_Moved"]/melted_HPI["Total_Settled"]
melted_HPI["Grad_Bach_Ratio"]=melted_HPI["Total_Grad"]/melted_HPI["Total_Bach"]
melted_HPI["College_Ratio"]=(melted_HPI["Total_Bach"]+melted_HPI["Total_Grad"])/melted_HPI["Total_pop"]
melted_HPI["Grad_Ratio_Moved"]=melted_HPI["Moved_Grad"]/melted_HPI["Total_Moved"]
melted_HPI["Years_Edu_Moved_Minus_Settled"]=melted_HPI["Years_Education_Moved"]-melted_HPI["Years_Education_Settled"]
melted_HPI["Total_Grad_Ratio"]=melted_HPI["Total_Grad"]/melted_HPI["Total_pop"]
melted_HPI["Total_Bach_Ratio"]=melted_HPI["Total_Bach"]/melted_HPI["Total_pop"]
melted_HPI.head()
```
