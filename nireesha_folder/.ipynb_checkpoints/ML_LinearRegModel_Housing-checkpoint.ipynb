```python
%matplotlib inline


import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score

import seaborn as sns
from math import sqrt
```

```python
training_df = pd.read_csv("MLResources/realestate_all_train.csv")
training_df.head()
training_df.count()
```

```python
plt.plot(training_df['Year'], training_df['Price'],'o')
```

```python
missing_df = training_df.loc[training_df['Year'].isin(['2003','2004'])]

missing_df.head()                 
```

```python
missing_2003_df = training_df.loc[training_df['Year'].isin(['2002'])]
missing_2003_df.dtypes

missing_2003_df.loc[:,'Year'] = 2003
```

```python
missing_2003_df.head()
```

```python
missing_2004_df = training_df.loc[training_df['Year'].isin(['2005'])]
missing_2004_df.dtypes

missing_2004_df.loc[:,'Year'] = 2004
```

```python
missing_2004_df.head()
```

```python
training_df = training_df.append(missing_2003_df, ignore_index=True)
```

```python
training_df = training_df.append(missing_2004_df, ignore_index=True)
```

```python
plt.plot(training_df['Year'], training_df['Price'],'o')
```

```python
print(training_df.loc[training_df['Price'].isin(['NaN'])])
```

```python
training_dfdum = training_df.dropna()
training_dfdum.head()
training_dfdum.count()
```

```python
print(training_dfdum.loc[training_dfdum['Price'].isin(['NaN'])])
```

```python
cat_features = ['County','State']
num_features = ['Price','Year']

categories = cat_features[0:len(cat_features)]

for category in categories:
    series = training_dfdum[category]
    dummies = pd.get_dummies(series, prefix=category)
    training_dfdum = pd.concat([training_dfdum, dummies], axis=1)
    
print(training_dfdum.columns)
training_dfdum.head(20)

# training_dfdum.count()
```

```python
removefeats= ["County","State","Price"]

from sklearn.preprocessing import StandardScaler
standardScalerX = StandardScaler()
x = standardScalerX.fit_transform(training_dfdum.drop(removefeats, axis=1))
x
```

```python
X_train = x
y_train = training_dfdum['Price']

regr = linear_model.LinearRegression()

lr_model = regr.fit(X_train, y_train)

y_train_predict = lr_model.predict(X_train)
print(f"True output: {y_train[0]}")
print(f"Predicted output: {y_train_predict[0]}")
print(f"Prediction Error: {y_train_predict[0]-y_train[0]}")
```

```python
removefeats= ["County","State","Price"]

X_train = training_dfdum.drop(removefeats, axis=1)
y_train = training_dfdum['Price']
```

```python
regr = linear_model.LinearRegression()

lr_model = regr.fit(X_train, y_train)
```

```python
print('Weight coefficients: ', lr_model.coef_)
print('y-axis intercept: ', lr_model.intercept_) 
```

```python
y_train_predict = lr_model.predict(X_train)
print(f"True output: {y_train[1]}")
print(f"Predicted output: {y_train_predict[1]}")
print(f"Prediction Error: {y_train_predict[1]-y_train[1]}")
```

```python
y_pred = lr_model.predict(X_train)
```

```python
lr_r2 =  r2_score(y_train, y_pred)
```

```python
bx=plt.subplots(figsize=(12,8))
bx= sns.barplot(x=0, y=1, data=pd.DataFrame(zip(X_train.columns, lr_model.coef_)))
plt.xticks(rotation='vertical')
plt.xlabel("Model Coefficient Types")
plt.ylabel("Model Coefficient Values")
plt.show()
print ("R squared: ", lr_r2)
print ("Average Coefficients: ", abs(lr_model.coef_).mean())
print ("Root Mean Squared Error: ", sqrt(mean_squared_error(y_train, y_pred)))
ax = sns.regplot(y_train, y_pred)
```

```python

train_accuracy = accuracy_score(y_train, y_train_predict)
# test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
# print('The test accuracy is', test_accuracy)
```
